```python
import pandas as pd
import pyarrow.parquet as pq
```

```python
taxi_jan_2024 = pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2024-01.parquet')
taxi_jan_2024.head(n=10)
```

```python
taxi_jan_2024.to_csv('yellow_tripdata_2024-01.csv', index=False)
```

```python
taxi_jan_2024.tpep_pickup_datetime = pd.to_datetime(taxi_jan_2024.tpep_pickup_datetime)
taxi_jan_2024.tpep_dropoff_datetime = pd.to_datetime(taxi_jan_2024.tpep_dropoff_datetime)
```

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://root:root@localhost:5433/ny_taxi')
```

```python
print(pd.io.sql.get_schema(taxi_jan_2024, name='yellow_taxi_data', con=engine))
```

```python
df_iter = pd.read_csv('yellow_tripdata_2024-01.csv', iterator=True, chunksize=100000)
df_iter
```

```python
taxi_jan_2024 = next(df_iter)
taxi_jan_2024
```

```python
taxi_jan_2024.tpep_pickup_datetime = pd.to_datetime(taxi_jan_2024.tpep_pickup_datetime)
taxi_jan_2024.tpep_dropoff_datetime = pd.to_datetime(taxi_jan_2024.tpep_dropoff_datetime)
```

```python
taxi_jan_2024.head(n=0).to_sql(name ='yellow_taxi_data',con=engine, if_exists='replace')
```

```python
%time taxi_jan_2024.to_sql(name ='yellow_taxi_data',con=engine, if_exists='append')
```

```python
from time import time
```

```python
while True:
    
    t_start = time()
    
    taxi_jan_2024 = next(df_iter)
    
    taxi_jan_2024.tpep_pickup_datetime = pd.to_datetime(taxi_jan_2024.tpep_pickup_datetime)
    taxi_jan_2024.tpep_dropoff_datetime = pd.to_datetime(taxi_jan_2024.tpep_dropoff_datetime)
    
    taxi_jan_2024.to_sql(name ='yellow_taxi_data',con=engine, if_exists='append')
    
    t_end = time()
    
    print('inserted another chunk, took %.3f seconds' % (t_end - t_start))
```
